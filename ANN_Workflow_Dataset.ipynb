# 2. import Dataset

```python
df = pd.read_csv("/content/drive/MyDrive/ANN/Churn_Modelling.csv")
# All the columns except last one
x = df.iloc[:, 3:-1].values
# Except all the columns only last column needed
y = df.iloc[:, -1].values
```

# 3. Data Pre-processing

1. Encoding Categorical data

Label encoding "Gender" column

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
x[:, 2] = le.fit_transform(x[:, 2])
```

```python
x
```

OneHotEnoding for "Geography" column

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [1])], remainder='passthrough')
x = np.array(ct.fit_transform(x))
```

```python
x
```

2. Spliting the dataset for train and test

```python
from sklearn.model_selection import train_test_split
x_train,x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, train_size=0.7, random_state=0)
```

3. Feature scaling


```python
# Feature Scaling for neural network
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

# 4. Building ANN

```python
# Initialize the ANN
ann = tf.keras.models.Sequential()

# Adding input layer and first hidden layer
ann.add(tf.keras.layers.Dense(units=6, activation='relu'))

# Adding second hidden layer
ann.add(tf.keras.layers.Dense(units=6, activation='relu'))

# Adding the output layer
ann.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))
```

# 5. Training ANN

1. Compiling the ANN

```python
ann.compile(optimizer = 'adam',
            loss = 'binary_crossentropy',
            metrics = ['accuracy'])
```

Training the ANN on training set

```python
# training the data
history = ann.fit(
    x_train,y_train,
    batch_size=32,
    epochs=100,
    verbose=True
)
```

**Important note-1:**  Notice that the values of the features were all input in a double pair of square brackets. That's because the "predict" method always expects 2D arrays as the format of its input and putting our values into a double pair of square brackets makes the input excatly a 2D array.

**Important note-2:** "france" column was not a input as a string in the last column but as "1, 0, 0" in the first three columns. Thats because of course the predict method expects the one-hot-encoded values of the state, and as we see in the first row of the matrix of the features X, "france" was encoded as "1, 0, 0" and be careful to include these values in the first three columns, because the dummy variables are always created in the first columns.

## Predicting the Test set results

```python
y_pred = ann.predict(x_test)
y_pred = (y_pred > 0.5)
print(np.concatenate((y_pred.reshape(len(y_pred), 1), y_test.reshape(len(y_test), 1)),1))
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```
